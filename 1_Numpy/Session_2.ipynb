1. Create a 2D NumPy array representing the ratings (out of 5) given by 4 users to 5 different food items on Zomato. Use slicing to extract the ratings given by the second and third users only.


```python
np.random.seed(42)
# row = user 
# column = 5 food items 
ratings = np.random.randint(0,6,20).reshape(4,5)
print(ratings)
## -- > ratings[[1,2],] this is using fancy indexing
ratings[1:3,] #--> this is slicing
```



2. Given a NumPy array of daily steps tracked for 10 days, use boolean indexing to select only the days where the steps were greater than 8000.

        Hint: Use an array like steps = np.array([7500, 8200, 9000, ... ]) and apply a boolean condition.


```python
np.random.seed(48)
steps = np.random.randint(0,10000,10)
print(steps)

np.where(steps > 8000)
# now i will give this to a list of user and can get user names with 8000+ steps 
```
