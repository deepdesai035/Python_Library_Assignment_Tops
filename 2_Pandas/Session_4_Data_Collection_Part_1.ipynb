
4. Download a large Excel file (at least 10,000 rows) of Flipkart product listings. Use pd.read_excel() with the chunksize parameter to read the file in chunks of 2000 rows, and print the number of rows in each chunk as you
iterate.


### Interview Question

Why does read_csv() support chunksize but read_excel() doesn't?

CSV files are plain text and can be streamed efficiently.
Excel files (.xlsx) are compressed workbook formats that usually need to be loaded before parsing.
Therefore, Pandas provides chunksize for CSVs but not for Excel files.

```python
dataset4 = pd.read_excel("F:/Tops_assignment/Python_Library/datasets/flipkart_product_excel/flipkart_product_listings_10000.xlsx")
dataset4

chunk_size = 2000

for i in range(0 , len(dataset4) ,chunk_size):
    chunk = dataset4.iloc[i:i+chunk_size]
    print(f"Chunk {(i//chunk_size) + 1 }: {len(chunk)} row")
```
