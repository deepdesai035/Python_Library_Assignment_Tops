4. Simulate a Zomato-style restaurant ratings dataset with some missing ratings. Use forward fill (method='ffill') to fill missing values in the ratings column, then use backward fill (method='bfill') for any remaining missing
values. Show the before and after results.


Forward Fill (ffill) → takes the previous valid value and fills downward.

Backward Fill (bfill) → takes the next valid value and fills upward.

Using both ensures most missing values are filled.

```python
dataset2 = pd.read_csv("F:/Tops_assignment/Python_Library/datasets/tsv_zomato/zomato_restaurants.tsv" , sep = '\t')
print(dataset2)

print("\n ------- \n \n")

dataset2 = dataset2.ffill()
print(dataset2)
```

```python
## now forward fill was not able to fill the first staring value which was null ,
# we will use backword fill it will fill the net value

dataset2 = dataset2.bfill()
print(dataset2)
```

5. Pick any one column with missing values from your dataset and explain which missingness mechanism (MCAR, MAR, or MNAR) is most likely and why, using 2-3 sentences.

        Hint: Think about whether the missing data is random or related to other factors in the dataset.



MCAR (Missing Completely At Random) → Missing values occur purely by chance.
Example: Data entry error randomly affects some rows.

MAR (Missing At Random) → Missing values depend on another observed column.
Example: New restaurants have more missing ratings.

MNAR (Missing Not At Random) → Missing values depend on the missing value itself.
Example: Restaurants with very poor ratings choose not to display them.
